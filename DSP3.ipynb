```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# 
def design_butterworth_lowpass(cutoff_freq, order):
    b, a = signal.butter(order, cutoff_freq, btype='low')
    return b, a

# 
def design_butterworth_highpass(cutoff_freq, order):
    b, a = signal.butter(order, cutoff_freq, btype='high')
    return b, a

# 
def design_butterworth_bandpass(low_cutoff, high_cutoff, order):
    b, a = signal.butter(order, [low_cutoff, high_cutoff], btype='band')
    return b, a

# 
def design_butterworth_bandstop(low_cutoff, high_cutoff, order):
    b, a = signal.butter(order, [low_cutoff, high_cutoff], btype='bandstop')
    return b, a

# 40.2 0  1
order = 2
cutoff_frequency = 0.2
b, a = design_butterworth_lowpass(cutoff_frequency, order)

# 
w, h = signal.freqz(b, a)
plt.plot(w/np.pi, 20 * np.log10(abs(h)), label='Lowpass')

# 40.2 0  1
b, a = design_butterworth_highpass(cutoff_frequency, order)
w, h = signal.freqz(b, a)
plt.plot(w/np.pi, 20 * np.log10(abs(h)), label='Highpass')

# 40.30.5 0  1
low_cutoff = 0.3
high_cutoff = 0.5
b, a = design_butterworth_bandpass(low_cutoff, high_cutoff, order)
w, h = signal.freqz(b, a)
plt.plot(w/np.pi, 20 * np.log10(abs(h)), label='Bandpass')

# 40.30.5 0  1
b, a = design_butterworth_bandstop(low_cutoff, high_cutoff, order)
w, h = signal.freqz(b, a)
plt.plot(w/np.pi, 20 * np.log10(abs(h)), label='Bandstop')

plt.title('Butterworth Filter Frequency Response')
plt.xlabel('Frequency [radians / sample]')
plt.ylabel('Amplitude [dB]')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# 
def design_butterworth_lowpass(cutoff_freq, order):
    b, a = signal.butter(order, cutoff_freq, btype='low')
    return b, a

# 
def design_chebyshev_lowpass(cutoff_freq, order, rp):
    b, a = signal.cheby1(order, rp, cutoff_freq, btype='low')
    return b, a

# 
order = 4
cutoff_frequency = 0.2  # 
rp = 3  # 

# 
b_butter, a_butter = design_butterworth_lowpass(cutoff_frequency, order)

# 
b_cheby, a_cheby = design_chebyshev_lowpass(cutoff_frequency, order, rp)

# 
w, h_butter = signal.freqz(b_butter, a_butter)
w, h_cheby = signal.freqz(b_cheby, a_cheby)

plt.figure(figsize=(10, 5))

# 
plt.subplot(2, 1, 1)
plt.plot(w/np.pi, 20 * np.log10(abs(h_butter)), label='Butterworth')
plt.plot(w/np.pi, 20 * np.log10(abs(h_cheby)), label='Chebyshev')
plt.title('Magnitude Response')
plt.xlabel('Frequency [radians / sample]')
plt.ylabel('Amplitude [dB]')
plt.legend()

# 
plt.subplot(2, 1, 2)
plt.plot(w/np.pi, np.angle(h_butter), label='Butterworth')
plt.plot(w/np.pi, np.angle(h_cheby), label='Chebyshev')
plt.title('Phase Response')
plt.xlabel('Frequency [radians / sample]')
plt.ylabel('Phase [radians]')
plt.legend()

plt.tight_layout()
plt.show()
```
